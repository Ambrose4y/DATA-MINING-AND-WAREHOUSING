**Cheking Regression Analysis**

```python
from sklearn.linear_model import LinearRegression
```

```python
ln = LinearRegression()
ln
```

```python
x = df2[['Personel']]
y = df2['Total cost']
```

```python
ln.fit(x,y)
```

**Prediction**

```python
Yhat=ln.predict(x)
Yhat[0:5]
```

**The intercept**

```python
ln.intercept_
```

The coefficent

```python
ln.coef_
```

**The estimated Linear Regression Model**

Total cost = 3,359,388. 07 + 2.27*(Personel)

**Multiple Linear Regression**

```python
Z = df2[['Personel', 'Transportation', 'Equipment',
       'General supplies', 'Building', 'Utilities', 'Curriculum']]
```

```python
ln.fit(Z, df2['Total cost'])
```

**The intercept**

```python
ln.intercept_
```

```python
ln.coef_
```

**The multiple Linear Regression Model**

Total cost = 69,158.97 + 1.49*(personel) + 0.23*(Transportation) + 0.197*(Equipment) - 0.002*(General supplies) + 0.52*(Building) + 0.81*(Utilities) + 0.99*(curriculum)

```python
Y_hat=ln.predict(Z)
Y_hat[0:5]
```

**In-sample evaluation**

**Mean square Error**

```python
from sklearn.metrics import mean_squared_error
```

**For simple linear regression**

**R^2**

```python
ln.fit(x, y)
print('The R-square is: ', ln.score(x, y))
```

```python
Yhat=ln.predict(x)
print('The output of the first four predicted value is: ', Yhat[0:4])
```

**Importing the mean squared error**

```python
mse=mean_squared_error(df2['Total cost'], Yhat)
print('The mean square error of price and predicted value is: ', mse)
```

**Multiple Linear Regression**

```python
ln.fit(Z, df2['Total cost'])
print('The R-square is: ', ln.score(Z, df2['Total cost']))
```

**MSE**

```python
mse=mean_squared_error(df2['Total cost'], Y_hat)
print('The mean square error of price and predicted value is: ', mse)
```

**The multiple linaer regression model in this case is the most appropriate to use in determining the Total Cost of educating a registered nurse in kenya since itsv r^2 value is 100% and Mse is very low.**
